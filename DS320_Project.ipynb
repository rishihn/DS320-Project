````python
import pandas as pd

imdb = pd.read_csv('imdb_top_1000.csv')
imdb.head()
````

````python
import pandas as pd

movies = pd.read_csv('top_4000_movies_data.csv')
movies.head()
````

Clean up and only highlight relevant info for imdb data set

````python
#imdb dataset: drop unneccesary columns and any NaN values
imdb_clean = imdb.drop(columns=['Poster_Link', 'Meta_score', 'Certificate', 'Overview', 'Star1', 'Star2', 'Star3', 'Star4'])
imdb_clean = imdb_clean.dropna()

#Change runtime from 142 min to just the integer
imdb_clean['Runtime'] = imdb_clean['Runtime'].str.extract('(\d+)').astype(int)

#Remove any commas or special characters from gross column
imdb_clean['Gross'] = imdb_clean['Gross'].replace('[\$,]', '', regex=True)

#Lowercase movie titles and remove any special characters like ':'
imdb_clean['Series_Title'] = imdb_clean['Series_Title'].replace('[\$,:]', '', regex=True)
imdb_clean['Series_Title'] = imdb_clean['Series_Title'].str.lower()

#rename Series_Title to just title
imdb_clean = imdb_clean.rename(columns={'Series_Title': 'title'})

#return cleaned dataset
imdb_clean
````

````python
import numpy as np

#movies dataset: drop unneccesary columns and any NaN values
movies_clean = movies.drop(columns=['rank', 'url', 'release_date', 'theaters'])
movies_clean = movies_clean.dropna()

#Convert year float values to simple integers
movies_clean['year'] = movies_clean['year'].astype(int)

#Convert runtime float values to simple integers
movies_clean['runtime'] = movies_clean['runtime'].astype(int)

#Convert opening_weekend float values to simple integers
movies_clean['opening_weekend'] = movies_clean['opening_weekend'].astype(int)

#Lowercase movie titles and remove any special characters like ':'
movies_clean['title'] = movies_clean['title'].replace('[\$,:]', '', regex=True)
movies_clean['title'] = movies_clean['title'].str.lower()

movies_clean
````
